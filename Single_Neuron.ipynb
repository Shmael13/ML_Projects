# Single Neuron Class 

---
Within this class, we will implement a single neuron. 
By applying different cost functions and preactivation functions to this class, we will be able to 
easily implement Logistic Regression, Linear Regression, and Perceptrons.

After Implementing this class, I will show how to use it with examples

Lastly, we will create a package so that the three types of machine learning (Logistic, Linear, Perceptron), can easily be used within other python programs.

---

```python
import matplotlib.pyplot as plt
import numpy as np

class Single_Neuron(object):
    def __init__(self, activation_function, cost_function):
        self.activation_function = activation_function
        self.cost_function = cost_function

    def train(self, X, y, epochs=1_000, learning_rate=0.001):
        self.weights_ = np.random.rand(1 + X.shape[1])
        self.errors_ = []
        N = X.shape[0]
        
        for _ in range(epochs):
            errors = 0
            for xi, ideal_target in zip(X, y):
                err = self.predict(xi) - ideal_target
                self.weights_[:-1] -= learning_rate * err * xi
                self.weights_[-1] -= learning_rate * err
                print(self.cost_function(0,1))
                errors += self.cost_function(self.predict(xi), ideal_target)
            self.errors_.append(errors/N)
        return self

    def predict(self, xi):
        preactivation = xi @ self.weights_[:-1] + self.weights_[-1]
        return self.activation_function(preactivation)

    def plot_inputs_vs_outputs(self, X, y):
        
        if(len(X.shape) > 2 or (X.shape[1] != 1 or X.shape[1] != 2) or y.shape[1] != 1):
            raise ValueError("Invalid Shape for Inputs or Outputs entered!")
        
        if (X.shape[1] == 1):
            plt.figure(figsize = (10,8))
            plt.title("Inputs vs Outputs")
            plt.xlabel("Inputs")
            plt.ylabel("Outputs")
            plt.plot(X, y)
            plt.show()
        if (X.shape[1] == 2):
            fig = plt.figure()
            ax = plt.axes(projection='3d')
            ax.set_title("Inputs vs Outputs - 3d")
            ax.set_xlabel("Inputs-#1")
            ax.set_ylabel("Inputs-#2")
            ax.plot3D(X[:, 0], X[:, 1], y[:, 0])
        return
    
            
```

# Activation Functions

```python
def linear_regression_activation(x):
    return x

def sign_activation(z):
    return np.sign(z)

def sigmoid_activation(z):
    return 1.0 / (1.0 + np.exp(-z))  
```

# Cost Functions

```python
def mean_squared_error(predicted, actual):
    return 0.5 * (predicted - actual) ** 2

def cross_entropy_loss(predicted, actual):
    return -actual * np.log(predicted) - (1 - actual)*np.log(1 - predicted)
```

```python
X = df[["sepal_length", "sepal_width"]].values
y = df.species.values
y = np.where(y == 'setosa', 1, -1)
print(y)
print(X)
```

```python
logistic_neuron = Single_Neuron(sigmoid_activation, cross_entropy_loss)
y = y.reshape(-1,1)
print(y.shape)
print(X.shape)
logistic_neuron.train(X, y)
y = logistic_neuron.predict(X)
```
